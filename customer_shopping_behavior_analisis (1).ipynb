```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))  #fill the null values in Review Rating
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()  # converting column names into lowercase string
df.columns=df.columns.str.replace(' ','_')  # replace space with '_'
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# creating age group using qcut in pandas
label=['young adult','adult','middle aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=label)  # qcut will cut age into 4 equal parts
```

```python
df[['age','age_group']].head(10) #head(10) means first 10 rows
```

```python
frequency_mapping={'Fortnightly':14,
                   'Weekly':7,
                   'Annually':365,
                   'Monthly':30,
                   'Quarterly':90,
                   'Bi-Weekly':14,
                   'Every 3 Months':90
                  }
df['purchase_frequency_days',]=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# checking whether 'Discount Applied' and 'Promo Code Used' has same value if yes then remove one of the column
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```
